```python
import json
import pickle
import time
import requests
from urllib.parse import urljoin
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
import pandas as pd
from bs4 import BeautifulSoup
```

```python
HOME_URL = 'https://www.bachhoaxanh.com/'
sidebar = BeautifulSoup(requests.get(HOME_URL).content, 'html.parser')
categories = sidebar.find('aside').find_all('a', href=True, class_='menu-haschild')
sub_links = [a['href'] for a in categories]
labels = [a.text[:-1].strip() for a in categories]
options = Options()
options.add_argument('--headless')
```

```python
print(sub_links)
```

```python
print(labels)
```

```python
all_parsed_products = {}
for i, link in enumerate(sub_links):
    driver = webdriver.Firefox(options = options)
    link = urljoin(HOME_URL, link)
    driver.get(link + '#trang-99')
    menu = driver.find_element_by_class_name('cate').get_attribute('innerHTML')
    driver.close()
    
    soup = BeautifulSoup(menu, 'html.parser')
    if (soup.find('div', class_='product-name') is not None):
        products = soup.find_all('div', class_='product-name')
    else:
        products = soup.find_all('h3', class_='product-name')
    keys = [t.text for t in products]
    values = [t.text[:-1] for t in soup.find_all('strong')]
    json_product = dict(zip(keys, values))
    all_parsed_products[labels[i]] = json_product
    print(json_product)
```

```python
f = open('./data/processed/all_products.bin', 'wb')
pickle.dump(all_parsed_products, f)
f.close()
```
